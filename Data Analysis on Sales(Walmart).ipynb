```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Walmart_Sales.csv')

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("\nCorrelation Matrix:")
print(df.corr())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Weekly_Sales'], marker='o', linestyle='-', color='b')
plt.title('Weekly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Weekly Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.histplot(df['Weekly_Sales'], kde=True)
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title('Weekly Sales by Holiday')
plt.xlabel('Holiday Flag (0 = No, 1 = Yes)')
plt.ylabel('Weekly Sales ')
plt.show()
```

```python
monthly_sales = df.resample('M', on='Date').sum()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales.index, monthly_sales['Weekly_Sales'], marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title('Temperature vs Weekly Sales')
plt.xlabel('Temperature (°F)')
plt.ylabel('Weekly Sales ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Fuel_Price', y='Weekly_Sales', data=df)
plt.title('Fuel Price vs Weekly Sales')
plt.xlabel('Fuel Price ($)')
plt.ylabel('Weekly Sales ($)')
plt.grid(True)
plt.show()
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
monthly_sales_mean = df.groupby('Month')['Weekly_Sales'].mean()
print(monthly_sales_mean)
```

```python
plt.figure(figsize=(10, 5))
monthly_sales_mean.plot(kind='bar', color='skyblue')
plt.title('Average Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Average Weekly Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

```python
holiday_sales = df[df['Holiday_Flag'] == 1]['Weekly_Sales']
non_holiday_sales = df[df['Holiday_Flag'] == 0]['Weekly_Sales']
print(holiday_sales,non_holiday_sales )
```

```python
print(f"Mean sales on holidays: ${holiday_sales.mean():.2f}")
print(f"Mean sales on non-holidays: ${non_holiday_sales.mean():.2f}")
print(f"Sales increase on holidays: {((holiday_sales.mean() - non_holiday_sales.mean()) / non_holiday_sales.mean()) * 100:.2f}%")
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='Month', y='Weekly_Sales', data=df)
plt.title('Weekly Sales by Month')
plt.xlabel('Month')
plt.ylabel('Weekly Sales ($)')
plt.show()

# Boxplot of Weekly Sales by Year
plt.figure(figsize=(14, 7))
sns.boxplot(x='Year', y='Weekly_Sales', data=df)
plt.title('Weekly Sales by Year')
plt.xlabel('Year')
plt.ylabel('Weekly Sales ($)')
plt.show()
```
